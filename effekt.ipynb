## Teorien der forklarer forsøget
Når energi omsættes fra én form til en anden, sker det ikke øjeblikkeligt — det tager tid. I forsøget ovenfor har du målt, hvor meget energi de forskellige apparater omsætter inden for et givent tidsrum. Det fører os frem til en ny fysisk størrelse: effekt.

Effekt er et udtryk for, hvor hurtigt energi omsættes, altså energi omsat pr. tidsenhed. Jo større effekt et apparat har, desto hurtigere omsætter det energi fra én form til en anden. Det er derfor effekten af en elkedel er langt større end effekten af en mobiltelefonoplader — elkedlen skal omsætte meget energi på kort tid for at varme vandet op, mens opladeren omsætter sin energi langsommere.

>**Effekt**
>
>Matematisk defineres effekten *P* som energien *E* divideret med tiden *t*, det tager at omsætte energien:
>
>$P = \frac{E}{t}$​ 

Da energi måles i Joule (J) og tid i sekunder (s), måles effekt i J/s. Denne enhed har fået sit eget navn: watt (W), opkaldt efter ingeniøren James Watt. Der gælder altså:

> **Enheden watt**
>
>$1 \text{W}=\frac{\text{J}}{\text{s}}$​

Grafisk svarer effekten til hældningen af (*t*,*E*)-grafen, når energien *E* afsættes som funktion af tiden *t* i et koordinatsystem. Hvis energien vokser jævnt med tiden — altså hvis sammenhængen er lineær — vil hældningen af grafen netop fortælle dig, hvor mange Joule der omsættes pr. sekund. En stejl graf betyder en stor effekt (energien omsættes hurtigt), mens en flad graf betyder en lille effekt (energien omsættes langsomt).

I forsøget kan du derfor bestemme effekten af hvert apparat ved at lave en lineær regression på dine målinger og aflæse hældningskoefficienten $a$. Denne hældning er apparatets effekt, målt i W, og den fortæller dig direkte, hvilket apparat der omsætter energi hurtigst.

```python
"""
(t, E)-graf for flere apparater
--------------------------------
Dette program tegner energi E som funktion af tiden t for flere apparater
med forskellig (konstant) effekt, samlet i ét koordinatsystem.

Da effekten for hvert apparat er konstant, er sammenhængen mellem E og t
lineær: E(t) = P * t, hvor hældningen netop er apparatets effekt P (i watt).

Jo stejlere linje, jo større effekt — og dermed jo hurtigere omsættes energien.

Kør programmet med: python3 apparater_effekt.py
"""

import numpy as np
import matplotlib.pyplot as plt

# ----------------------------------------------------------
# 1. Definér apparater og deres effekt (i watt)
# ----------------------------------------------------------
apparater = {
    "LED-pære":                 {"effekt": 8,    "farve": "#16a34a"},
    "Glødepære":                 {"effekt": 60,   "farve": "#f59e0b"},
    "Opladning af mobiltelefon": {"effekt": 10,   "farve": "#9333ea"},
    #"Elkedel":                   {"effekt": 2000, "farve": "#dc2626"},
}

t_max = 60  # sekunder, samme tidsrum som i forsøget
t = np.linspace(0, t_max, 100)

# ----------------------------------------------------------
# 2. Opsæt figur med to grafer:
#    - venstre: alle apparater (inkl. elkedel)
#    - højre: kun de tre svageste apparater (så man kan se forskellen)
# ----------------------------------------------------------
fig, ax1 = plt.subplots(1, 1, figsize=(8, 6))

for navn, data in apparater.items():
    P = data["effekt"]
    farve = data["farve"]
    E = P * t  # E(t) = P * t, da effekten er konstant

    ax1.plot(t, E, label=f"{navn} (P = {P} W)", color=farve, linewidth=2.5)

# ----------------------------------------------------------
# 3. Formatér graf 
# ----------------------------------------------------------
ax1.set_xlabel("Tid (s)", fontsize=12)
ax1.set_ylabel("Energi (J)", fontsize=12)
ax1.set_title("Energiomsætning med forskellig effekt", fontsize=13, fontweight="bold")
ax1.grid(True, alpha=0.3)
ax1.set_xlim(0, t_max)
ax1.set_ylim(0, None)
ax1.legend(loc="upper left", fontsize=9)

#plt.tight_layout()
#plt.savefig("apparater_effekt.png", dpi=150, bbox_inches="tight")
plt.show()
```

### Hvordan beregner man energi og tid?
Lad os tage udgangspunkt i sammenhængen mellem effekt, energi og tid:

$P = \frac{E}{t}$​

Denne formel kan også omskrives, så vi kan finde enten energien eller tiden, hvis vi i stedet kender effekten.

#### Eksempel 1 - Beregning af energi og tid
Forestil dig en LED-pære med en effekt på 10 W. Det betyder, at den omsætter 10 J energi for hvert sekund, den er tændt.

**Hvor meget energi omsætter LED-pæren, hvis den er tændt i 2 sekunder?**

På 2 sekunder omsætter LED-pæren dobbelt så meget energi som på 1 sekund. Det giver derfor god mening at gange de to tal sammen:

$E = 10 \text{J/s} \cdot 2 \text{s} = 20 \text{J}$

Energien stiger altså, jo længere tid apparatet kører — det er præcis det, du så som en ret linje i forsøget med de fire apparater.

**Hvor lang tid tager det for LED-pæren at omsætte 40 J?**

Her kender vi energien og effekten, men ikke tiden. Da effekten fortæller os, at der omsættes 10 Joule på 1 sekund, må det tage 4 sekunder at omsætte 40 J. Det kan vi beregne ved at dividere:

$t = \frac{40 \text{J}}{10 \text{J/s}} = 4 \text{s}$

#### Formlerne
De to beregninger ovenfor er ikke tilfældige — de følger direkte af at isolere *E* og *t* i den oprindelige formel $P = \frac{E}{t}$

Energien finder du ved at gange effekten med tiden:

$E = P \cdot t$

Det giver mening, fordi effekten netop angiver, hvor mange Joule der omsættes for hvert sekund — så for at finde den samlede energi, skal du blot gange op med, hvor mange sekunder apparatet har kørt.

Tiden finder du ved at dividere energien med effekten:

$t = \frac{E}{P}$​

Her spørger du i virkeligheden: "Hvis apparatet omsætter *P* Joule hvert sekund, hvor mange sekunder tager det så at nå op på de *E* Joule, jeg har målt?"

>Husk enhederne: Energien skal være i Joule (J), effekten i watt (W) eller Joule pr. sekund ($\text{J/s}$) og tiden i sekunder (s), for at formlerne stemmer overens. Er effekten i stedet angivet i kW, og tiden i timer, ender man i stedet med energien i kWh — se afsnittet om kWh.

## Uddybning af teorien

### Energienheden kWh
I dagligdagen, og især når man taler om elregninger, måler man sjældent energi i Joule — tallene bliver hurtigt enorme. I stedet bruges enheden **kilowatttime (kWh)**. Navnet afslører selv, hvordan enheden er bygget op: den er et udtryk for energi, fundet ved at gange en effekt (i kilowatt) med en tid (i timer). Det giver god mening, når man husker på, at $E = P \cdot t$ — det er bare den omvendte formel af den, vi brugte til at finde effekten i forsøget.

Da 1 kW = 1000 W = 1000 J/s, og 1 time = 3600 s, kan man omregne mellem de to energienheder:

> **Enheden kilowatttimer (kWh)**
>
>$1 \text{kWh}=1000 \text{W}\cdot 3600 \text{s} = 3.600.000 \text{J}=3,6 \text{MJ}$


Når man f.eks. ser, at en elkedel med en effekt på 2 kW har kørt i en halv time, kan man hurtigt regne energiforbruget ud som $E = 2 \text{kW} \cdot 0,5 \text{h} = 1\text{kWh}$ — uden at skulle regne om til Joule og sekunder undervejs. Det er denne enhed, elselskaberne bruger til at afregne dit elforbrug, og den effekt, dine apparater er mærket med (f.eks. "7 W" på en pære), fortæller dig præcis, hvor hurtigt de omsætter energi, og dermed hvor meget det koster at have dem tændt.

### Varierende effekt

I praksis er effekten af et apparat ikke altid konstant. Tag for eksempel en elkedel: lige når den tændes, er effekten muligvis lidt lavere, og når vandet nærmer sig kogepunktet, kan effekten ændre sig igen. Når effekten varierer over tid, kan man ikke længere bruge den simple formel $P = \frac{E}{t}$​ til at beskrive hele forløbet — den formel giver kun **gennemsnitseffekten** i det tidsrum, man måler over.

Grafisk ses dette ved, at grafen for energi som funktion af tid ikke længere er en ret linje, men en kurve. Hældningen af en kurve er ikke konstant, men ændrer sig undervejs. Det betyder, at effekten på et bestemt tidspunkt svarer til hældningen af **tangenten** til kurven i det punkt — altså den øjeblikkelige stigningstakst for energien lige dér. Man taler her om **momentan effekt**, i modsætning til den gennemsnitlige effekt, man finder ved at dividere den samlede energi med den samlede tid.

Når effekten er konstant, er sammenhængen mellem energi og tid lineær, og så er den momentane effekt og gennemsnitseffekten naturligvis den samme — det er derfor, lineær regression giver god mening i forsøget med de fire apparater. Men for mange apparater i hverdagen — f.eks. en støvsuger, der bruges i ryk, eller en bil, der accelererer og bremser — varierer effekten konstant, og man må i stedet se på effekten time for time eller endda sekund for sekund.
